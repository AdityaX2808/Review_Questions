```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
spark = SparkSession.builder.appName("Week7").getOrCreate()
df = spark.read.option("header" , True).option("inferSchema" , True).csv("D:\projects\BL_DE_Digit Insurance\ReviewProblems\week7\Car Sales.xlsx - car_data.csv")
df.show()
```

```python
df.printSchema()
df.describe().show()
```

for each company each body style all products avg price resale wise.

```python
df_result = df.groupBy("Company","Body Style").agg(avg("Price ($)").alias("average_price"))
```

```python
import matplotlib.pyplot as plt
df_plot = df_result.toPandas()
plt.bar(df_plot["Company"] , df_plot["average_price"])
plt.xlabel("Company")
plt.ylabel("Average Price")
plt.xticks(rotation = 90 )
plt.show()
```

```python
plt.bar(df_plot["Body Style"] , df_plot["average_price"])
plt.xlabel("body Style")
plt.ylabel("Average Price")
plt.xticks(rotation = 90 )
plt.show()
```
